# Data Visualization With Matplotlib
Matplot is a plotting library for creating static, animated and interactive visualizations!!

## Creating a Simple Plot:
You can create various types of plots easily.

```python
# Example
import matplotlib.pyplot as plt
```

```python
x = [1, 2, 3, 4]
y = [10, 15, 7, 10]
```

```python
plt.plot(x, y)
plt.title('Simple Matplotlib Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show();
```
